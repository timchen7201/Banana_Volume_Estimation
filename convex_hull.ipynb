```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial.qhull import ConvexHull
from scipy.spatial.distance import euclidean
```

```python
img=cv2.imread("images/edged_img/image_03-07-2020_01-00-05_AM.png")
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
value, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
img=thresh
```

```python
img.shape
```

```python
img1=img[50:img.shape[0]-50,50:img.shape[0]-50]
plt.imshow(img1)
```

```python
img1
```

```python
img1.shape
```

```python
tmp=img1.copy
contours, hierarchy = cv2.findContours(tmp,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
```

```python
points = img1
hull = ConvexHull(points)

vertices = hull.vertices.tolist() + [hull.vertices[0]]
perimeter = np.sum([euclidean(x, y) for x, y in zip(points[vertices], points[vertices][1:])])
print(perimeter)
```
